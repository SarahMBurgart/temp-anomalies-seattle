```python
import pandas as pd
from matplotlib import pyplot as plt

#make it pretty
plt.style.use('ggplot')
```

```python
# https://www.ncdc.noaa.gov/paleo-search/study/10437 1000 year 
# 1,000 Year Ensemble Reconstructions of Temperature and CO2

# calibration period - 1971- 2000

df = pd.read_csv("/Users/sarahburgart/galvanize/week4 - capstone 1/data/shi2013nh.txt",sep="\t", low_memory=False, comment='#')
df.head()
```


A temperature anomaly is the difference from an average, or baseline, temperature. 
The baseline temperature is typically computed by averaging 30 or more years of temperature data.

```python
fig, ax = plt.subplots(1, 1, figsize=(24, 8))
ax.plot(df["age_AD"],df["tempanom-ann"], label = "annual temperature anomalies")
ax.plot(df["age_AD"],df["tempanom-ann.1"], label = "annual temperature anomalies-1")
ax.plot(df["age_AD"],df["tempanom-ann.2"], label = "annual temperature anomalies-2")
ax.plot(df["age_AD"],df["tempanom-ann.3"], label = "annual temperature anomalies-3s")

#ax.set_yticks()

#ax.set_xticks()
ax.set_title("tempanom-ann")
ax.legend(loc='best')

plt.show()


```
